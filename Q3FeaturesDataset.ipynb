```python
import pandas as pd

#Features Dataset
df=pd.DataFrame({
    'ID':['M001','M002','M003','M004','M005','M006'],
    'Feature1':[5.1,4.9,6.2,5.9,5.6,5.7],
    'Feature2':[3.5,3.0,3.4,3.0,2.9,2.8],
    'Feature3':[1.4,1.4,5.4,5.1,3.6,4.1],
    'Feature4':[0.2,0.2,2.3,1.8,1.3,1.3],
    'Label':['Setosa','Setosa','Virginica','Virginica','Versicolor','Versicolor']
}
)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Features and target
X = df[["Feature1","Feature2","Feature3","Feature4"]]
y = df["Label"]

# Train-test split 
X_train = X
y_train = y

# KNN model
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

# Predictions
pred = model.predict(X_train)

print(pred)
```

```python
#Euclidean distances
from scipy.spatial.distance import pdist, squareform

features = df[["Feature1","Feature2","Feature3","Feature4"]]

dist_matrix = pd.DataFrame(
    squareform(pdist(features, metric="euclidean")),
    index=df["ID"],
    columns=df["ID"]
)

print(dist_matrix)
```

```python
#Standardizing data to apply PCA
from sklearn.preprocessing import StandardScaler

X = df[["Feature1","Feature2","Feature3","Feature4"]]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```
